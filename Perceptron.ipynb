```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def activation(x):
  return 0 if(x<0) else 1

def predict(input,weights,bias):
  output = np.dot(input, weights) + bias
  return activation(output)

def perceptron(inputs,weights,bias,learning_rate,targets,epochs):
  for i in range(epochs):
    for input,target in zip(inputs,targets):
      output=predict(input,weights,bias)
      error=target-output
      weights+=learning_rate*error*input
      bias+=learning_rate*error
  return weights,bias
```

```python
def decision_boundary(inputs, outputs, weights, bias):
  colors = ['blue' if output else 'red' for output in outputs]
  plt.figure(figsize=(6,6))
  plt.scatter(inputs[:, 0], inputs[:, 1], c=colors, s=100, edgecolors='k', label='Data Points (Colored by Prediction)')
  x_values = np.linspace(-0.1, 1.1, 200)
  y_values = -(weights[0] * x_values + bias) / weights[1]

  plt.plot(x_values, y_values, 'g--', label="Decision Boundary")

  plt.xlabel("Input 1")
  plt.ylabel("Input 2")
  plt.title("Perceptron Decision Boundary")
  plt.xlim(-0.5, 1.5)
  plt.ylim(-0.5, 1.5)
  plt.grid(True)
  plt.legend()
  plt.show()
```

**AND Gate**

```python
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
targets = np.array([0, 0, 0, 1])
weights = np.array([0.1,0.1])
bias = 0.1
learning_rate=0.01
epochs=100

print(f"Initial Weights: {weights}, Initial Bias: {bias}")
weights, bias = perceptron(inputs,weights,bias,learning_rate,targets,epochs)
print(f"Updated Weights: {weights}, Updated Bias: {bias}")
```

```python
print("\nInput - Output")
outputs = [predict(input, weights, bias) for input in inputs]
for input, output in zip(inputs, outputs):
    print(f"{input} - {output}")
```

```python
decision_boundary(inputs, outputs, weights, bias)
```

**OR Gate**

```python
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
targets = np.array([0, 1, 1, 1])
weights = np.array([0.1,0.1])
bias = 0.1
learning_rate=0.01
epochs=100

print(f"Initial Weights: {weights}, Initial Bias: {bias}")
weights, bias = perceptron(inputs,weights,bias,learning_rate,targets,epochs)
print(f"Updated Weights: {weights}, Updated Bias: {bias}")
```

```python
print("\nInput - Output")
outputs = [predict(input, weights, bias) for input in inputs]
for input, output in zip(inputs, outputs):
    print(f"{input} - {output}")
```

```python
decision_boundary(inputs, outputs, weights, bias)
```
