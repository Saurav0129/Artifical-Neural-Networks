**Weight and Bias effect on Output.**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def calculate_output(inputs, weights, bias):
    return sigmoid(np.dot(inputs, weights) + bias)

inputs = np.linspace(-10, 10, 100)
true_weights = 2.0
true_bias = 1.0
true_output = calculate_output(inputs, true_weights, true_bias)
```

```python
weights_range = np.linspace(1, 3, 5)
for weight in weights_range:
    output = calculate_output(inputs, weight, true_bias)
    plt.plot(inputs, output, label=f'Weight: {weight:.2f}')

plt.plot(inputs, true_output, label='True Output', linestyle='--')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Effect of Varying Weights on Output')
plt.legend()
plt.show()
```

```python
bias_range = np.linspace(1, 3, 5)
for bias in bias_range:
    output = calculate_output(inputs, true_weights, bias)
    plt.plot(inputs, output, label=f'Bias: {bias:.2f}')

plt.plot(inputs, true_output, label='True Output', linestyle='--')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Effect of Varying Bias on Output')
plt.legend()
plt.show()
```

**Conclusion**

The plots illustrate the effects of varying weights and biases on the output of a simple neural network with a sigmoid activation function.

1. **Effect of Varying Weights:**
   - When varying weights while keeping the bias constant, we observe that the shape of the output curve changes.
   - Increasing the weights tends to amplify the input signal, leading to a steeper curve with higher output values, while decreasing the weights flattens the curve and lowers the output values.
   - This demonstrates how the network learns to amplify or diminish the importance of different features in the input data based on the weights assigned to them.

2. **Effect of Varying Bias:**
   - When varying biases while keeping the weights constant, we observe shifts in the output curve along the y-axis.
   - Increasing the bias shifts the curve upwards, while decreasing the bias shifts it downwards.
   - This illustrates how biases allow the network to adjust the overall output of the activation function, helping the model to better fit the data by providing flexibility in output values.

In conclusion, the weight parameter controls the strength of connections, influencing the timing and amplitude of activation, while the bias parameter modulates the threshold for activation, governing the level at which activation occurs.
