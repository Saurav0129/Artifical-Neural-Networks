**Activation functions to train Neural Network**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exp_vals = np.exp(x - np.max(x))
    return exp_vals / np.sum(exp_vals)

def linear(x):
    return x

def hard_limiter(x, threshold=0.5):
    return np.where(x > threshold, 1, 0)
```

```python
X = np.array([[0.1, 0.2],
              [0.2, 0.3],
              [0.4, 0.5]])

W = np.array([[0.5, 0.3],
              [0.2, 0.4]])

b = np.array([0.1, 0.2])
```

```python
def forward(X, W, b, activation_func):
    return activation_func(np.dot(X, W) + b)
```

```python
print("Sigmoid Activation:")
print(forward(X, W, b, sigmoid))
print("\nReLU Activation:")
print(forward(X, W, b, relu))
print("\nTanh Activation:")
print(forward(X, W, b, tanh))
print("\nSoftmax Activation:")
print(forward(X, W, b, softmax))
print("\nLinear Activation:")
print(forward(X, W, b, linear))
print("\n Hard Limiter Activation:")
print(forward(X, W, b,hard_limiter))
```

```python
x = np.linspace(-5, 5, 100)
sigmoid_y = sigmoid(x)
relu_y = relu(x)
tanh_y = tanh(x)
softmax_y = softmax(x)
linear_y = linear(x)
hard_limiter_y=hard_limiter(x)

plt.figure(figsize=(12,10))
plt.plot(x, sigmoid_y, label='Sigmoid', color='blue')
plt.plot(x, relu_y, label='ReLU', color='red')
plt.plot(x, tanh_y, label='Tanh', color='green')
plt.plot(x, softmax_y, label='Softmax', color='purple')
plt.plot(x,hard_limiter_y,label="Hard Limiter(threshold=0.5)",color='pink')
plt.plot(x, linear_y, label='Linear', color='orange', linestyle='--')
plt.title('Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```

# Conclusion

1. **Sigmoid**: Sigmoid function squashes the input values between 0 and 1, useful for binary classification problems where you need to output probabilities.

2. **ReLU (Rectified Linear Unit)**: ReLU function returns 0 for negative inputs and the input itself for positive inputs, widely used in deep learning due to its simplicity and effectiveness in mitigating the vanishing gradient problem.

3. **Tanh**: Tanh function squashes the input values between -1 and 1, similar to sigmoid but with output mean centered around 0, often used in recurrent neural networks (RNNs).

4. **Softmax**: Softmax function is used for multi-class classification problems to produce a probability distribution over multiple classes, ensuring that the output probabilities sum up to 1.

5. **Linear**: Linear activation function simply outputs the input value without any transformation, commonly used in the output layer for regression tasks.

6. **Hard Limiter (Threshold)**: Hard limiter function clips the input values to be either below or above a specified threshold, useful for binary decision-making scenarios where outputs are binary.


Introducing non-linear activation functions like ReLU, sigmoid, tanh, and softmax enables neural networks to learn complex patterns in data. These functions introduce non-linearity, allowing networks to capture intricate relationships between input features and output predictions, ultimately enhancing performance across various tasks.
